```python
data = pd.read_csv('140828_tweet_activity_metrics.csv')
#data.head()
```

```python
er = data['engagement rate']
imp = data['impressions']
tm = data['time']
```

```python
import re
def time_of_day_factory():
    pattern = re.compile(r"^([0-9]*)-([0-9]*)-([0-9]*)\ ([0-9]*):([0-9]*)\ ([+-])(.*)$")
    def f(s):
        m = pattern.match(s)
        mg = m.groups(); 
        return float(mg[3]) + 0.01666667 * float(mg[4])
    return f

time_of_day = time_of_day_factory()
```

```python
tod = list(map(time_of_day, tm))
```

```python
plt.plot(tod, er, '+')
plt.ylim([0,1])
plt.title("Engagement Rate vs Time of Day")
```

```python
plt.plot(tod, er, '+')
plt.ylim([0,0.2])
plt.title("Engagement Rate vs Time of Day")
```

```python
plt.hist(tod, bins=100)
plt.title("Number of Tweets vs Time of Day")
```
